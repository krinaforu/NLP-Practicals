```python
import pandas as pd
data = pd.read_json("Quora_data2.json")
data.to_csv("Quora_answers.csv")
data.head()
```

```python
df= data[['answers']]
df.head()
```

```python
df = df.replace('\n','',regex=True)
df.fillna("Not Available")
```

#### Tokenizing

```python
from nltk.tokenize import word_tokenize, RegexpTokenizer
```

```python
words =[]
for i in df["answers"]:
    tokenizer = RegexpTokenizer(r'\w+')
    word = tokenizer.tokenize(i)
    words.extend(word)
words = [x.lower() for x in words]
words
```

```python
len(words)
```

```python
#Calculating frequency
from nltk import FreqDist
frequency = FreqDist(words)
frequency
```

#### Lemmatizing

```python
from nltk import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lem_word = []
for i in words:
    lem_word.append(lemmatizer.lemmatize(i))
lem = [lem_word]
```

#### Creating N-Gram

```python
from nltk import bigrams,trigrams
from collections import Counter,defaultdict
```

#### Unigram model

```python
uni = {w1: 0 for i in lem for w1 in i}
```

```python
#Counting frequency
for sentence in lem:
    for w1 in sentence:
        uni[w1] += 1
print(uni)
```

```python
total_count = float(sum(uni.values()))
for w1 in uni:
    uni[w1] /= total_count
uni
```

```python
model_uni = {k: v for k,v in sorted(uni.items(),key=lambda item :item[1],reverse=True)}
model_uni
```

#### Bigram model

```python
big = list(bigrams(lem_word))
big
```

```python
model_bigr = defaultdict(lambda: defaultdict(lambda:0))
```

```python
#Counting frequency
for sentence in lem:
    for w1,w2 in bigrams(sentence,pad_right=True, pad_left=True):
        model_bigr[(w1)][w2] += 1
```

```python
#Estimating probability
for w1 in model_bigr:
    total_count = float(sum(model_bigr[w1].values()))
    for w2 in model_bigr[w1]:
        model_bigr[w1][w2] /= total_count
```

```python
dict(model_bigr["tea"])
```

#### Trigram model

```python
tri = list(trigrams(lem_word))
tri
```

```python
model_trigr = defaultdict(lambda: defaultdict(lambda:0))
```

```python
#Counting frequency
for sentence in lem:
    for w1,w2,w3 in trigrams(sentence,pad_right=True, pad_left=True):
        model_trigr[(w1,w2)][w2] += 1
```

```python
#Estimating probability
for w1_w2 in model_trigr:
    total_count = float(sum(model_trigr[w1_w2].values()))
    for w3 in model_trigr[w1_w2]:
        model_trigr[w1_w2][w3] /= total_count
```

```python
dict(model_trigr["white","tea"])
```

#### Finding random probability

```python
from copy import deepcopy as dc
```

```python
def findprob(m1,m2,text):
    text=text.split()
    text.insert(0,None)
    prob=1
    m1[None]=1
    for i in range(1,len(text)):
        if text[i] not in m1:
            temp=(m2[text[i-1]][text[i]]+1)/(len(m1))
            prob*=temp
        else:
            temp=(m2[text[i-1]][text[i]]+1)/(m1[text[i]]+len(m1))
            prob*=temp
    return prob
```

```python
text="black tea"
findprob(dc(model_uni),dc(model_bigr),text)
```
