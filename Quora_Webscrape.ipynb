```python
import json
import ssl
import urllib
from urllib.request import Request,urlopen
from bs4 import BeautifulSoup
import requests
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
url = input("Enter the question that you want to find answer for -\n")
url = "https://www.quora.com/" + url.replace(" ","-")
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.content, "html.parser")
Quora_json=dict()
```

```python
question = soup.find("title")
Quora_json["question"]=question.text.replace("- Quora"," ")
```

```python
answers = soup.find("script",{"type":"application/ld+json"})
ans_list = json.loads(answers.string)#["mainEntity"]["suggestedAnswer"]
ans_list
```

```python
for number, answer in enumerate(ans_list["mainEntity"]["suggestedAnswer"], start=1):
    print(f"Answer: {number}. | Upvote count: {answer['upvoteCount']}")
    print(answer["text"].strip())
    print("-" * 75)
```

```python
#print(type(ans_list["mainEntity"]["suggestedAnswer"]))
#for answer in ans_list["mainEntity"]["suggestedAnswer"][0]:
#    print(answer)
#print(ans_list["mainEntity"]["suggestedAnswer"][1]["text"])
Quora_json["answers"]=[]
for i in range(0,10):
    answer_val= {"name": ans_list["mainEntity"]["suggestedAnswer"][i]["author"]["name"],
                 "text": ans_list["mainEntity"]["suggestedAnswer"][i]["text"],
                 "upvoteCount": ans_list["mainEntity"]["suggestedAnswer"][i]["upvoteCount"]}
   # answer_val_list.append(answer_val)
    Quora_json["answers"].append(answer_val)
```

```python
print(Quora_json)
```

```python
with open("Quora_data.json",'w')as outfile:
    json.dump(Quora_json,outfile,indent=4)
```
